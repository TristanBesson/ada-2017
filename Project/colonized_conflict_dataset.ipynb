```python
import pandas as pd
import requests
import re
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium
import os
import branca.colormap as cm
```

```python
clean_conflict = pd.read_csv('datasets/clean_conflict.csv')
colonisation_df = pd.read_csv('datasets/colonies_wikipedia.csv')
colonisation_df = colonisation_df.replace("Britain", "United Kingdom")
```

```python
# We need to add several column: If the country was colonised, the year of decolonisation and the country that colonised.
colonizer_country = colonisation_df['Colonizer Country'].tolist()
colonized_country = colonisation_df['Colonized Country'].tolist()
year = colonisation_df['Year'].tolist()
sidea = clean_conflict['sidea'].tolist()
location  = clean_conflict['location'].tolist()
terr = clean_conflict['terr'].tolist()
```

```python
#colonized_conflict = []
year_dec_conflict = []
colonizer_conflict= []
id_drop = []
```

```python
for i in range(0, len(terr)):
    if terr[i] in colonizer_country:
        j = colonizer_country.index(terr[i])
       #colonized_conflict.append(colonized_country[j])
        year_dec_conflict.append(year[j])
        colonizer_conflict.append(colonizer_country[j])
    elif location[i] in colonizer_country:
        j = colonizer_country.index(location[i])
        #colonized_conflict.append(colonized_country[j])
        year_dec_conflict.append(year[j])
        colonizer_conflict.append(colonizer_country[j])
    elif sidea[i] in colonizer_country:
        j = colonizer_country.index(sidea[i])
        #colonized_conflict.append(colonized_country[j])
        year_dec_conflict.append(year[j])
        colonizer_conflict.append(colonizer_country[j])
    else:
        #colonized_conflict.append(0)
        year_dec_conflict.append(0)
        colonizer_conflict.append(0)
        id_drop.append(i)
```

```python
len(colonizer_conflict)
```

```python
clean_conflict = clean_conflict.assign(decolonisation_year = year_dec_conflict )
clean_conflict = clean_conflict.assign(colonizer = colonizer_conflict )

clean_conflict = clean_conflict.drop(clean_conflict.index[id_drop])
```

```python
clean_filename = 'datasets/clean_conflict_of_colonized_countries.csv'

clean_conflict.to_csv(clean_filename)
```
